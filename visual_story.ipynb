```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

plt.rcParams["figure.figsize"] = (8,5)
```

```python
df = pd.read_csv("../data/titanic.csv")

df.head()
```

```python
# Fill missing Age values
df['Age'] = df['Age'].fillna(df['Age'].median())

# Fill missing Embarked values
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Drop Cabin column
df.drop(columns=['Cabin'], inplace=True)
```

```python
sns.countplot(data=df, x='Survived', palette='Set2')

plt.title("Overall Survival Count")

plt.xlabel("Survived (0 = No, 1 = Yes)")

plt.ylabel("Number of Passengers")

plt.show()
```

**Takeaway:** More passengers died than survived during the Titanic disaster.

```python
sns.countplot(data=df, x='Sex', hue='Survived', palette='coolwarm')

plt.title("Survival by Gender")

plt.xlabel("Gender")

plt.ylabel("Passenger Count")

plt.show()
```

**Takeaway:** Female passengers had much higher survival rates than male passengers.


```python
sns.countplot(data=df, x='Pclass', hue='Survived', palette='viridis')

plt.title("Survival by Passenger Class")

plt.xlabel("Passenger Class")

plt.ylabel("Passenger Count")

plt.show()
```

**Takeaway:** First-class passengers survived more frequently than third-class passengers.

```python
sns.histplot(df['Age'], bins=30, kde=True, color='skyblue')

plt.title("Age Distribution of Passengers")

plt.xlabel("Age")

plt.ylabel("Frequency")

plt.show()
```

**Takeaway:** Most Titanic passengers were young adults between ages 20 and 40.

```python
sns.histplot(df['Fare'], bins=40, color='orange')

plt.title("Fare Distribution")

plt.xlabel("Fare")

plt.ylabel("Frequency")

plt.show()
```

**Takeaway:** Most passengers paid relatively low fares, while a few paid very high ticket prices.

```python
sns.countplot(data=df, x='Embarked', hue='Survived', palette='magma')

plt.title("Survival by Embarkation Port")

plt.xlabel("Embarkation Port")

plt.ylabel("Passenger Count")

plt.show()
```

**Takeaway:** Survival patterns differed across embarkation ports due to passenger demographics and class distribution.

# Summary

This analysis explored survival patterns in the Titanic dataset using visual storytelling techniques. The charts showed that survival was strongly linked to gender and passenger class. Female passengers survived at much higher rates than male passengers, while first-class passengers had significantly better survival chances compared to lower classes.

The age distribution revealed that most passengers were young adults, while the fare distribution showed large differences in passenger wealth. Higher ticket fares were associated with improved survival rates. Survival patterns also varied across embarkation ports, reflecting differences in passenger demographics and travel class.

Several cleaning steps were completed before visualization. Missing Age values were replaced using the median age, missing Embarked values were filled using the most common category, and the Cabin column was removed because it contained too many missing values.

Overall, the project demonstrates how visualization can clearly explain the relationship between social status, gender, and survival during the Titanic disaster.
